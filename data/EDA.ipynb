```python
numerical_columns = ['age', 'chol', 'trestbps', 'thalach', 'oldpeak']
categorical_columns = ['sex', 'cp', 'fbs', 'restecg', 'slope', 'ca', 'thal']
```

```python
for col in categorical_columns:
    print(f"\nUnique values in {col}:")
    print(df[col].value_counts())
```

```python
from scipy.stats import skew, kurtosis
for col in numerical_columns:
    print(f"\nSkewness of {col}: {skew(df[col].dropna()):.2f}")
    print(f"Kurtosis of {col}: {kurtosis(df[col].dropna()):.2f}")
```

Age:
Skewness (-0.24): Slightly negatively skewed, indicating a small tail towards younger ages but generally symmetric.
Kurtosis (-0.53): Platykurtic distribution, meaning it has lighter tails and a flatter peak compared to a normal distribution.

Cholesterol (chol):
Skewness (1.09): Positively skewed, suggesting a longer tail towards higher cholesterol levels.
Kurtosis (4.07): Leptokurtic distribution, meaning it has heavier tails and a sharper peak than a normal distribution, indicating outliers or extreme values.

Resting Blood Pressure (trestbps):
Skewness (0.73): Moderately positively skewed, indicating higher values have a longer tail.
Kurtosis (0.96): Slightly leptokurtic, suggesting mild outliers but relatively normal distribution.

Maximum Heart Rate Achieved (thalach):
Skewness (-0.52): Slightly negatively skewed, showing a small tail towards higher heart rate values.
Kurtosis (-0.09): Mesokurtic or close to normal distribution, suggesting typical variability without extreme outliers.

ST Depression (oldpeak):
Skewness (1.22): Positively skewed, indicating a significant tail towards higher ST depression values.
Kurtosis (1.35): Leptokurtic, suggesting the presence of significant outliers or extreme values in the dataset.

Summary:
The variables chol and oldpeak exhibit strong positive skewness and high kurtosis, which indicate potential outliers that should be examined further.
Age and thalach are relatively symmetric and close to normal distributions, making them less problematic for analysis.
Trestbps shows moderate skewness, requiring potential transformation if strict normality is needed for modeling.

```python
for col in numerical_columns:
    plt.figure()
    sns.boxplot(x=df[col], palette='Set3')
    plt.title(f'Outliers in {col}')
    plt.xlabel(col)
    plt.show()
```

```python
for col in categorical_columns:
    cross_tab = pd.crosstab(df[col], df['output'])
    cross_tab.div(cross_tab.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, figsize=(8, 6), colormap='viridis')
    plt.title(f'Stacked Bar Chart of {col} by output')
    plt.xlabel(col)
    plt.ylabel('Proportion')
    plt.show()
```

```python
correlations = df.corr()['output'].sort_values(ascending=False)
print(f"\nCorrelations with output:\n")
print(correlations)
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Heatmap')
plt.show()
```
